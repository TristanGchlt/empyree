```python

from collections import Counter
```

```python
corpus_path = "../data/processed/card_corpus.txt"
with open(corpus_path, "r", encoding="utf-8") as f:
    decks = [line.strip().split() for line in f]

print(f"Nombre de decks : {len(decks)}")
deck_lengths = [len(deck) for deck in decks]
print(f"Min/Max/Moy cartes par deck : {min(deck_lengths)}, {max(deck_lengths)}, {sum(deck_lengths)/len(deck_lengths):.2f}")
```

```python
# Taille du vocabulaire
vocab = set(card for deck in decks for card in deck)
print(f"Taille du vocabulaire : {len(vocab)}")
```

```python
# Top 20 cartes
counter = Counter(card for deck in decks for card in deck)
print("Top 20 cartes les plus fr√©quentes :")
for card, freq in counter.most_common(20):
    print(card, freq)
```
